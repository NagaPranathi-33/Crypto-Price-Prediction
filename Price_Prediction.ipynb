```python
#LSTM--> Long Short Term Memory
#scale the data
#using minmaxscaler for scaling the data into the ranges [-1,1]

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(closingPrice[['Close']].dropna())
```

```python
scaled_data
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
#prepare data for LSTM
x_data = []
y_data = []
basedays = 100
for i in range(basedays, len(scaled_data)):
    x_data.append(scaled_data[i-basedays: i])
    y_data.append(scaled_data[i])
x_data = np.array(x_data)
y_data = np.array(y_data)
    
    
```

```python
#Split into train and test data
train_size = int(len(x_data) * 0.85)

x_train, y_train = x_data[:train_size], y_data[:train_size]
x_test, y_test = x_data[train_size:], y_data[train_size:]
```

```python
#building model..!

model = Sequential([
    LSTM(128, return_sequences = True, input_shape = (x_train.shape[1], 1)),
    LSTM(64, return_sequences = False), 
    Dense(25),
    Dense(1)
])

model.compile(optimizer = "adam", loss = "mean_squared_error")
model.summary()
```

```python
#model fitting..

model.fit(x_train, y_train, batch_size = 5, epochs = 8)
```

```python
#Start predictions..

predictions = model.predict(x_test)
inv_predictions = scaler.inverse_transform(predictions)
inv_y_test = scaler.inverse_transform(y_test)
```

```python
#the 3 columns are some values..
inv_predictions
```

```python
plotting_data = pd.DataFrame(
    {
        'Original': inv_y_test.flatten(),
        'Prediction':inv_predictions.flatten()
    }, index = closingPrice.index[train_size + basedays:]
)

```

```python
#Plot closing price with enhancement

plt.figure(figsize = (15, 6))
plt.plot(plotting_data.index, plotting_data['Original'], label = 'Original', color = 'blue', linewidth=1.5)
plt.plot(plotting_data.index, plotting_data['Prediction'], label = 'Prediction', color = 'red', linewidth=1.5)

plt.title("Prediction vs Actual Close Price", fontsize = 16)
plt.xlabel("Years", fontsize = 14)
plt.ylabel("Close Price", fontsize= 14)
plt.grid(alpha = 0.3)
plt.legend(fontsize = 12)
plt.show()
```

```python
#lets predict future days..

last_100 = scaled_data[-100:].reshape(1, -1, 1)
future_predictions = []
for _ in range(10):
    next_days = model.predict(last_100)
    future_predictions.append(scaler.inverse_transform(next_days))
    last_100 = np.append(last_100[:, 1:, :], next_days.reshape(1, 1, -1), axis = 1)
```

```python
future_predictions
```

```python
future_predictions = np.array(future_predictions).flatten()

#Plot closing price with enhancement

plt.figure(figsize = (15, 6))
plt.plot(range(1, 11), future_predictions, marker="o", label = 'Prediction future prices', color = 'purple', linewidth=1.5)

for i, val in enumerate(future_predictions):
    plt.text(i+1, val, f'{val:.2f}', fontsize = 10, ha = 'center', va = 'bottom', color = 'black')
    
plt.title("Future Close Prices for 10 days", fontsize = 16)
plt.xlabel("Day Ahead", fontsize = 14)
plt.ylabel("Close Price", fontsize= 14)
plt.grid(alpha = 0.3)
plt.legend(fontsize = 12)
plt.show()
```
