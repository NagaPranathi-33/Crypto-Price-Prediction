```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
from datetime import datetime

end = datetime.now()
start = datetime(end.year-15, end.month, end.day, end.hour, end.minute)
stock = 'BTC-USD'
stock_data = yf.download(stock, start=start, end=end)
```

```python
stock_data.head()
stock_data.tail()
```

```python
stock_data.describe().T
```

```python
stock_data.info()
```

```python
#opening price data
openingPrice = stock_data[['Open']]
openingPrice
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming 'OpeningPrice' is a DataFrame with 'Op' as a column and index as datetime
plt.figure(figsize=(15, 8))

# Plotting
plt.plot(
    openingPrice.index, openingPrice['Open'], 
    label='Open Price', color='blue', linewidth=1.5
)

# Enhancements
plt.title("Open Price of Bitcoin", fontsize=18, fontweight='bold', pad=20)
plt.xlabel("Years", fontsize=14, labelpad=10)
plt.ylabel("Open Price (USD)", fontsize=14, labelpad=10)
plt.grid(alpha=0.4, linestyle='--')  # Lighter and dashed grid
plt.legend(fontsize=12)

# Customize x-axis date ticks
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Show every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format as 'YYYY'
plt.xticks(rotation=45, fontsize=12)  # Rotate labels for readability
plt.yticks(fontsize=12)

plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```

```python
#close price data
closingPrice = stock_data[['Close']]
closingPrice
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming 'closingPrice' is a DataFrame with 'Close' as a column and index as datetime
plt.figure(figsize=(15, 8))

# Plotting
plt.plot(
    closingPrice.index, closingPrice['Close'], 
    label='Close Price', color='blue', linewidth=1.5
)

# Enhancements
plt.title("Close Price of Bitcoin", fontsize=18, fontweight='bold', pad=20)
plt.xlabel("Years", fontsize=14, labelpad=10)
plt.ylabel("Close Price (USD)", fontsize=14, labelpad=10)
plt.grid(alpha=0.4, linestyle='--')  # Lighter and dashed grid
plt.legend(fontsize=12)

# Customize x-axis date ticks
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Show every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format as 'YYYY'
plt.xticks(rotation=45, fontsize=12)  # Rotate labels for readability
plt.yticks(fontsize=12)

plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
